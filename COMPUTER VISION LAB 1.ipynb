 # IMAGE PROCESSING

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
image=cv2.imread('Downloads/lenna.png')
image
```

## **IMAGE READING**

```python
plt.imshow(image)
plt.axis('off')
```

## **DIMENSION OF THE IMAGE**

```python
height,width,channel=image.shape
print("height",height)
print("width",width)
print("channel",channel)
```

# BGR TO RGB CONVERSION

## USING THE INBUILT FUNCTION

```python
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image)
plt.axis('off')
```

## WITHOUT USING THE INBUILT FUNCTION(MANUAL)

```python
rgb_image = image[:, :,[2,1,0]]
plt.imshow(rgb_image)
plt.axis('off')
```

# BGR TO GRAY SCALE CONVERSION

## WITH USING THE INBUILT FUNCTION

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image,cmap='gray')
plt.axis("off")
```

## WITHOUT USING THE INBUILT FUNCTION

```python
gray_image = 0.2989 * image[:, :, 2] + 0.5870 * image[:, :, 1] + 0.1140 * image[:, :, 0]
plt.imshow(gray_image,cmap='gray')
plt.axis("off")
```

## INTENSITY OF THE IMAGES

```python

dark_image = np.clip(gray_image * 0.1, 0, 255).astype(np.uint8)  


bright_image = np.clip(gray_image * 1.5, 0, 255).astype(np.uint8) 

plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(gray_image, cmap='gray')
plt.axis('off')


plt.subplot(1, 3, 2)
plt.title('Darkened Image')
plt.imshow(dark_image, cmap='gray')
plt.axis('off')


plt.subplot(1, 3, 3)
plt.title('Brightened Image')
plt.imshow(bright_image, cmap='gray')
plt.axis('off')


plt.tight_layout()
plt.show()
```

## IMAGE RESIZING BY SETTING THE VALUE

```python
new_size = (80, 60) 
resized_image = cv2.resize(image, new_size)
```

```python
resized_image
```

##  **VISUALIZATION OF RESIZED IMAGE**



```python


plt.title("Original Image")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')



plt.title("Resized Image")
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()



```

## IMAGE RESIZING BY DIVIDING THE SIZE(WIDTH,HEIGHT)

```python
resized_image=cv2.resize(image,(int(image.shape[1]/5),int(image.shape[0]/4)))
plt.imshow(resized_image)
plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))                    
plt.axis('off')
resized_image
```

# HISTOGRAM EQUILIZATION


## USING INBUILT FUNCTION

```python

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
equalized_image = cv2.equalizeHist(gray_image)

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(gray_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Equalized Image')
plt.imshow(equalized_image, cmap='gray')
plt.axis('off')
plt.show()
```

## WITHOUT USING INBUILT FUNCTION EQUILIZATION

```python

histogram, bins = np.histogram(gray_image.flatten(), bins=256, range=[0, 256])
cdf = histogram.cumsum()
cdf_normalized = (cdf - cdf.min()) * 255 / (cdf.max() - cdf.min())

equalized_image = np.interp(gray_image.flatten(), bins[:-1], cdf_normalized).reshape(gray_image.shape)


plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(gray_image, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Equalized Image (Manual)')
plt.imshow(equalized_image, cmap='gray')
plt.axis('off')

plt.show()


```

## HISTPLOT FOR WITHOUT USING INBUILT FUNCTION EQUILIZATION

```python


gray_image = 0.2989 * image[:, :, 2] + 0.5870 * image[:, :, 1] + 0.1140 * image[:, :, 0]


hist, bins = np.histogram(gray_image.flatten(), bins=256, range=[0, 256])
cdf = hist.cumsum()
cdf_normalized = cdf * 255 / cdf.max()


equalized_image = np.interp(gray_image.flatten(), bins[:-1], cdf_normalized)
equalized_image = equalized_image.reshape(gray_image.shape)


plt.title('Original Grayscale Image')
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.show()


plt.title('Histogram of Original Image')
plt.plot(hist, color='blue')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()


plt.title('Histogram Equalized Image')
plt.imshow(equalized_image, cmap='gray')
plt.axis('off')
plt.show()


equalized_hist, _ = np.histogram(equalized_image.flatten(), bins=256, range=[0, 256])


plt.title('Histogram of Equalized Image')
plt.plot(equalized_hist, color='red')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

## HISTPLOT FOR USING INBUILT FUNCTION EQUILIZATION

```python


gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
hist = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
equalized_image = cv2.equalizeHist(gray_image)


equalized_hist = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])


plt.title('Original Grayscale Image')
plt.imshow(gray_image, cmap='gray')
plt.axis('off')
plt.show()


plt.title('Histogram of Original Image')
plt.plot(hist, color='blue')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()


plt.title('Histogram Equalized Image')
plt.imshow(equalized_image, cmap='gray')
plt.axis('off')
plt.show()


plt.title('Histogram of Equalized Image')
plt.plot(equalized_hist, color='red')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```
